```python
# CBTIS No. 128
# Submodulo 2: Elabora proyectos de programación lógica
# Docente: Eliseo Nava
# Alumno: Rafael Arevalo
# Especialidad: Inteligencia Artificial

# Clase Dentista
class Dentista:

    def __init__(self):
        self.dentistas = []

    # Agregar dentista
    def agregar_dentista(self):
        dentista = {}

        print("\n--- AGREGAR DENTISTA ---")

        dentista["nombre"] = input("Nombre: ")
        dentista["numero_colegiado"] = input("Número de colegiado: ")
        dentista["clinica_propia"] = input("Clínica propia (si/no): ")
        dentista["servicios_ofrecidos"] = input("Servicios ofrecidos: ")
        dentista["materiales_stock"] = input("Materiales en stock: ")
        dentista["citas_dia"] = input("Citas del día: ")
        dentista["anios_practica"] = input("Años de práctica: ")

        self.dentistas.append(dentista)

        print("Dentista agregado correctamente")

    # Listar dentistas
    def listar_dentistas(self):
        print("\n================ LISTADO DE DENTISTAS ================")

        if len(self.dentistas) == 0:
            print("No hay dentistas registrados")
        else:
            print("{:<12} {:<12} {:<10} {:<20} {:<12} {:<8} {:<8}".format(
                "Nombre", "Colegiado", "Clínica", "Servicios",
                "Materiales", "Citas", "Años"))
            print("-" * 95)

            for d in self.dentistas:
                print("{:<12} {:<12} {:<10} {:<20} {:<12} {:<8} {:<8}".format(
                    d["nombre"],
                    d["numero_colegiado"],
                    d["clinica_propia"],
                    d["servicios_ofrecidos"],
                    d["materiales_stock"],
                    d["citas_dia"],
                    d["anios_practica"]
                ))

    # Actualizar dentista
    def actualizar_dentista(self):
        nombre = input("Nombre del dentista a actualizar: ")

        for d in self.dentistas:
            if d["nombre"] == nombre:
                d["citas_dia"] = input("Nuevo número de citas: ")
                d["materiales_stock"] = input("Nuevo material en stock: ")
                print("Dentista actualizado")
                return

        print("Dentista no encontrado")

    # Borrar dentista
    def borrar_dentista(self):
        nombre = input("Nombre del dentista a borrar: ")

        for d in self.dentistas:
            if d["nombre"] == nombre:
                self.dentistas.remove(d)
                print("Dentista eliminado")
                return

        print("Dentista no encontrado")


# Crear objeto
sistema = Dentista()

# Menú principal
while True:
    print("\n=========== MENU ===========")
    print("1. Agregar dentista")
    print("2. Listar dentistas")
    print("3. Actualizar dentista")
    print("4. Borrar dentista")
    print("5. Salir")
    print("============================")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        sistema.agregar_dentista()

    elif opcion == "2":
        sistema.listar_dentistas()

    elif opcion == "3":
        sistema.actualizar_dentista()

    elif opcion == "4":
        sistema.borrar_dentista()

    elif opcion == "5":
        print("Programa terminado")
        break

    else:
        print("Opción inválida")
```
