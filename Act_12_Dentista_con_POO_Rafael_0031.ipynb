```python
# UIII Act 13 Proyecto de simulación de inteligencia artificial
# Rafael Arevalo

class Dentista:

    def __init__(self, nombre, colegiado, clinica, servicios, materiales, citas, anios):
        self.nombre = nombre
        self.colegiado = colegiado
        self.clinica = clinica
        self.servicios = servicios
        self.materiales = materiales
        self.citas = citas
        self.anios = anios

    def mostrar_datos(self):
        print("{:<12} {:<12} {:<10} {:<22} {:<12} {:<8} {:<8}".format(
            self.nombre,
            self.colegiado,
            self.clinica,
            self.servicios,
            self.materiales,
            self.citas,
            self.anios
        ))


dentistas = []


while True:
    print("\n=========== MENU ===========")
    print("1. Agregar dentista")
    print("2. Listar dentistas")
    print("3. Actualizar dentista")
    print("4. Borrar dentista")
    print("5. Salir")
    print("============================")

    opcion = input("Selecciona una opción: ")

    if opcion == "1":
        print("\n--- AGREGAR DENTISTA ---")

        nombre = input("Nombre: ")
        colegiado = input("Número de colegiado: ")
        clinica = input("Clínica propia (si/no): ")
        servicios = input("Servicios ofrecidos: ")
        materiales = input("Materiales en stock: ")
        citas = int(input("Citas del día: "))
        anios = int(input("Años de práctica: "))

        nuevo = Dentista(nombre, colegiado, clinica, servicios, materiales, citas, anios)
        dentistas.append(nuevo)

        print("Dentista agregado correctamente")

    elif opcion == "2":
        print("\n================ LISTADO DE DENTISTAS ================")

        print("{:<12} {:<12} {:<10} {:<22} {:<12} {:<8} {:<8}".format(
            "Nombre", "Colegiado", "Clínica", "Servicios",
            "Materiales", "Citas", "Años"
        ))

        print("-" * 95)

        for dentista in dentistas:
            dentista.mostrar_datos()

    elif opcion == "3":
        nombre_buscar = input("Nombre del dentista a actualizar: ")

        encontrado = False

        for dentista in dentistas:
            if dentista.nombre == nombre_buscar:
                dentista.citas = int(input("Nuevo número de citas: "))
                dentista.materiales = input("Nuevo material en stock: ")
                print("Dentista actualizado")
                encontrado = True
                break

        if not encontrado:
            print("Dentista no encontrado")

    elif opcion == "4":
        nombre_borrar = input("Nombre del dentista a borrar: ")

        eliminado = False

        for dentista in dentistas:
            if dentista.nombre == nombre_borrar:
                dentistas.remove(dentista)
                print("Dentista eliminado")
                eliminado = True
                break

        if not eliminado:
            print("Dentista no encontrado")

    elif opcion == "5":
        print("Programa terminado")
        break

    else:
        print("Opción inválida")
```
